8. Crea un gr치fico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sof칤a mide 190 cm, Carmen mide 170 cm. Las barras ser치n verdes.
Consejo: si no vas a tener que manipular especialmente los ejes, es m치s sencillo dejarlos dentro de la figura.

```python
x = ['Ana', 'Luis', 'Pedro', 'Sof칤a', 'Carmen']
y = [160, 180, 175, 190, 170]

plt.title("Altura de amigos")
plt.xlabel("Amigos")
plt.ylabel("Alturas")
plt.bar(x,y, color = "green");
```

9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los m칠todos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
barplot = plt.bar(x,y)

for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va = "bottom")


plt.title("Altura de amigos")
plt.xlabel("Amigos")
plt.ylabel("Alturas")
plt.bar(x,y, color = "green");
```

22. Contornos de 3D a 2D. Escribe una funci칩n que recibiendo x,y devuelva (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

Tablero: Tanto x como y van de -3 a 3 y usaremos 256 puntos.
Usa contourf, con 8 niveles (cortes), una transparencia de 0.75 y un color map de tipo 'jet'. 
Bonus: pinta las l칤neas de los contornos tambi칠n de negro

```python
def f(x,y):
    return (1-x / 2 + x ** 5 + y ** 3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X, Y = np.meshgrid(x,y)
plt.contourf(X, Y, f(X,Y), 8, alpha = .75, cmap = "jet")
C = plt.contour(X, Y, f(X, Y), 8, colors = 'black')
```
